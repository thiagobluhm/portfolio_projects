#### Capturando somente os csv's

```python
# Use glob para encontrar todos os arquivos .csv na pasta
arqs_csv = glob.glob(os.path.join('*.csv'))
```

#### Concatenando os dataframes com a lista criada acima

```python
df_ = pd.DataFrame()
for arq in arqs_csv:
    df = pd.read_csv(arq, sep=";")
    df_ = pd.concat([df, df_])
```

####  Reorganizando

```python
df__ = df_.sort_values('COD_UF', ascending=True)
```
